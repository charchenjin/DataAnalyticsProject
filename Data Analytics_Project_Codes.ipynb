## Random Forest

```python
# define independent attributes 
features = ['Time', 'V1', 'V2', 'V3', 'V4', 'V5',\
            'V6', 'V7', 'V8', 'V9', 'V10', 'V11',\
            'V12', 'V13', 'V14', 'V15', 'V16', 'V17',\
            'V18', 'V19','V20', 'V21', 'V22', 'V23',\
            'V24', 'V25', 'V26', 'V27', 'V28','Amount']

# assign values for independent variables and target variable

X = data [features]
y = data ['Class']
```

```python
# import train_test split
from sklearn.model_selection import train_test_split

# split train data into train and test, 60% in training and 40% in testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state = 0)
```

```python
#method1: oversampling 
#apply the resampling to the training set only

from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler()
X_ros, y_ros = ros.fit_resample(X_train, y_train)
```

```python
from collections import Counter

print(Counter(y))
print(Counter(y_ros))
```

```python
# import libararies for evaluation measures
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_oversample = RandomForestClassifier(random_state = 42)
rfc_oversample.fit(X_ros,y_ros)

pred_val = rfc_oversample.predict(X_test)

pred_prob = rfc_oversample.predict_proba(X_test)

print("Accuracy:",accuracy_score(pred_val, y_test, normalize=True, sample_weight=None))

# https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report
print("Classification Report:",classification_report(pred_val, y_test))

# https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html
print("Confusion Matrix:", confusion_matrix(pred_val, y_test))
```
